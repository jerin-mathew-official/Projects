# Demonstrate Naive Bayes Classifier

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
df = pd.read_csv('C:/ClassesMSC/Bank_Personal_Loan_Modelling.csv')
df.info()
```

```python
df.head()
```

```python
X = df.drop(columns=['Online'])
y = df.Online
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 1)
X_train.head()
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.ensemble import RandomForestClassifier

#using the random forest classifier for bagging and feature randomness to create uncorrelation

model = RandomForestClassifier()
model.fit(X_train,y_train)

#getting the predictions using the trained model

predictions = model.predict(X_test)

#printing the accuracy of the train data
print("the accuracy of the train data is ",model.score(X_train,y_train)*100)
                                                                                 
#printing the accuracy of the test data
print("the accuracy of the test data is",model.score(X_test,y_test)*100)
```

```python
#using quatile transformer to transform the dataset into a normal distribution

from sklearn.preprocessing import QuantileTransformer
qt = pd.DataFrame(QuantileTransformer(output_distribution='normal').fit_transform(X_train))
qt
```

```python
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results

y_pred = classifier.predict(X_test)
y_pred
```

```python
# Making the Confusion Matrix for finding the positive, false_positive, negative and false_negative
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
#Checking the accuracy score of the model

ac = (accuracy_score(y_test,y_pred)*100)
ac
```
