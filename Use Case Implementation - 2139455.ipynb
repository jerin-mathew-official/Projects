### Some Analysis and Findings on Wage:

#### .Wage is in String Data type.
#### .Some Wage values is in form amountK (Starting with  and end with K)
#### .All the values are not having K at the end.
#### .So at first we have to remove  and K and transform Wedge value into numeric value.

```python
df["Wage"].head()
```

```python
## Let us remove  and K

df.Wage = df.Wage.str.replace("","")
df.Wage = df.Wage.str.replace("K","").astype("float")
df.Wage.head() 
```

  ### Objective 2 : PREFFERED FOOT USED BY PLAYERS 

```python
preferred_foot = df["PreferredFoot"].value_counts()
preferred_foot = preferred_foot.reset_index()
preferred_foot.columns = ["PreferredFoot","Count"]
print(preferred_foot)
```

#### From analysis above, it has been found that 13948 playes use Right leg preferably and 4211 players use preferably Left leg. It can be observed from bar chart below.

```python
plt.bar(x = preferred_foot['PreferredFoot'], height = preferred_foot['Count'],
       color = ['DarkBlue', 'yellow'])
plt.xlabel('Preferred Foot')
plt.ylabel('Count')
plt.title('Distribution of Preferred Foot in FIFA 2019')
plt.show()
```

### Objective 3: DOES PREFFERED FOOT IS HAVING SOME EFFECT ON OVERALL RANKING ?

```python
sns.catplot(x="PreferredFoot", y="Overall", data=df,
            aspect=2, kind="bar", palette="cubehelix");
plt.style.use("ggplot")
plt.title("Figure 2 : \n\n Overall Rating vs Preffered Foot",
          fontsize=25);
```

#### From figure, It is not very clear that which preffered foot get more rating. So it requires some more statistical test analysis.

### Objective 4: BEST PLAYER IN EACH POSITION 

```python
temp_df = pd.DataFrame(columns = (df.columns))
grouped_fifa = df.groupby('Position')
for p, df in grouped_fifa:
    sort_df = df.sort_values('Overall', ascending = False).head(1)
    temp_df = temp_df.append(sort_df, ignore_index = True)
    
best_players = temp_df[['Name', 'Overall', 'Position']].sort_values('Position')
best_players
```

### Objective 5: HOW MANY PLAYERS HAVE 100K - 500K WAGE? 

```python
top_wage = df[(df['Wage'] >= 100) & (df['Wage'] <=500)]
print('There are',top_wage.Name.nunique(), 'players have the wage which is in the range 100k-500k')
```

### Objective 6 : AGE DISTRIBUTION OF FIFA PLAYERS

```python
age_df = df.Age.values
plt.figure(figsize = (12,6))
plt.hist(x = age_df,edgecolor = 'black', color ='blue')
plt.title('Age Distribution of FIFA Players (2019)', fontsize = 20)
plt.xlabel('Age', fontsize = 15)
plt.ylabel('Frequency', fontsize = 15)
plt.show()
```

### Objective 7: NATION WISE PLAYER COUNT OF TOP 20 COUNTRIES

```python
counts_Nationality = df["Nationality"].value_counts()
counts_Nationality = counts_Nationality.reset_index()
counts_Nationality.columns= ["Nations","Counts"]
counts_Nationality.head()
```

```python
Nationality_top20 = counts_Nationality.iloc[0:20,:]
print(Nationality_top20)
```

```python
sns.catplot(y="Nations",
            x="Counts",data=counts_Nationality_top20,
            palette="colorblind",height=9,kind="bar",aspect=2)
plt.title("Figure : \n\n\n Nation Wise Players Counts for top 20 country",
          fontsize=25);
```

### Objective 8: FINDING THE BEST CLUBS BASED ON PLAYER OVERALL 

```python
club = df[['Name','Overall','Club']]
club.head(10)
```

```python
bestclub = df.groupby('Club').Overall.mean().reset_index().sort_values(by='Overall', ascending=False)
bestclub_top10 = bestclub[:10]
plt.figure(figsize=(10,5))
sns.barplot(y=bestclub_top10['Club'], x=bestclub_top10.Overall, palette='YlGnBu',edgecolor='black')
plt.ylabel('Club', size=15)
plt.xlabel('Mean Rating', size=15)
plt.title('Best Clubs\n(Player\'s Overall Mean Rating)', size=25)
plt.show
```
