### 1. Generating sample dataset/use any open data set:
#### Here I will use the popular "Air Quality" dataset from the UCI Machine Learning Repository for this task. The dataset contains several features such as temperature, humidity, CO2 levels, and more, recorded hourly over several months.

```python
import seaborn as sns
import numpy as np
import pandas as pd
data = pd.read_csv("C:/ClassesMSC/DataSets/AirPassengers.csv",usecols=[1], engine='python', skipfooter=3)
```

### 2.|Preprocess data (reshaping):
#### I will reshape the data so that it can be used for training a RNN model..

```python
# Reshape data
data = data.values.reshape(-1, 1)
```

### 3.Build a Simple RNN model:
#### I will use the Sequential class from Keras to create a simple RNN model with one LSTM layer and one fully connected layer.

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Define the model
model = Sequential()
model.add(LSTM(4, input_shape=(1, 1)))
model.add(Dense(1))

# Compile the model
model.compile(loss='mean_squared_error', optimizer='adam')
```

### 4.Print or plot the prediction results:
#### I will use the fit method to train the model on the data and then use the predict method to generate predictions.

```python
# Fit the model
model.fit(data, data, epochs=100, batch_size=1, verbose=2)

# Generate predictions
predictions = model.predict(data)
```

#### I will plot the predictions and the original data together to visualize the results

```python
import matplotlib.pyplot as plt
plt.plot(data, label='Original data')
plt.plot(predictions, label='Predictions')
plt.legend()
plt.show()
```
